### Load libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Transcribe data
We are transcribing from Figure 6 of https://arxiv.org/pdf/2103.00020.pdf.

<img src="clip-paper-fig-6.png" width="400">

```python
n_trains = np.array([1, 2, 4, 8, 16])
height_in_pixels = 450
pixels_above_30 = np.array([146, 240, 324, 365, 416])
accs = (30 + 45 * (pixels_above_30 / height_in_pixels)) / 100
accs
```

```python
plt.plot(n_trains, 1 - accs, "o-")
plt.xlabel("Number of training images / class")
plt.ylabel("Error rate")
plt.xscale("log")
plt.yscale("log")
```

No cliffs to be seen when averaged across "20 datasets with at least 16 examples per class."
