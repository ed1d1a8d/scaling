```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs_up = np.arange(100) + 1
ys_up = xs_up ** -0.5

xs_down = np.arange(100000)
ys_down = (len(xs_up) + xs_down) ** -0.5

plt.plot(xs_up, ys_up, label="up")
plt.plot(xs_down, ys_down, label="down")
plt.xscale("log")
plt.yscale("log")
plt.legend()
```
